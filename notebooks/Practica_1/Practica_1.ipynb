## Previo de la Práctica 1

## 1. Definición de URDF

El **Unified Robot Description Format (URDF)** es un formato basado en XML utilizado en ROS (Robot Operating System) para describir modelos de robots. Este formato permite definir la estructura física del robot, especificando enlaces (links) y juntas (joints) que describen la configuración cinemática y dinámica del robot. El URDF también incluye información sobre la visualización y los modelos de colisión del robot. Los robots descritos con URDF deben tener una estructura de árbol, lo que excluye robots con componentes paralelos o flexibles&#8203;:contentReference[oaicite:0]{index=0}.

## 2. Tipos de Juntas en URDF

En URDF, las juntas definen cómo se conectan y mueven los enlaces de un robot. Los tipos de juntas disponibles son:

- **Revolute**: Permiten la rotación alrededor de un eje con un rango limitado.
- **Continuous**: Permiten rotación continua sin límites.
- **Prismatic**: Permiten deslizamiento a lo largo de un eje.
- **Fixed**: Conexiones rígidas sin movimiento.
- **Floating**: Permiten movimiento en los seis grados de libertad.
- **Planar**: Permiten movimiento en un plano perpendicular al eje de la junta&#8203;:contentReference[oaicite:1]{index=1}.

## 3. Definición de Elementos "Primitivos" en URDF

Los elementos primitivos del URDF son los componentes básicos que definen la estructura de un robot:

- **Link (Enlace)**: Describe los cuerpos rígidos del robot, incluyendo su geometría y modelos de colisión.
- **Joint (Junta)**: Define las conexiones y movimientos entre los enlaces.
- **Geometry**: Describe la forma de los enlaces usando primitivas como cajas, esferas, cilindros o mallas complejas importadas&#8203;:contentReference[oaicite:2]{index=2}.

## 4. Qué es un Nodo y un Tópico en ROS 2

- **Nodo**: Es un proceso ejecutable en ROS que realiza tareas específicas y se comunica con otros nodos mediante la publicación y suscripción de datos.
- **Tópico**: Es un canal de comunicación que permite a los nodos enviar y recibir mensajes, facilitando la transferencia de información entre procesos.

## 5. Qué es RViz

**RViz** es una herramienta de visualización para ROS que permite a los usuarios ver el estado de un robot y los datos sensoriales en un entorno 3D. Se utiliza principalmente para visualizar modelos URDF, sensores, trayectorias y otros elementos clave en la robótica, lo que facilita el desarrollo y la depuración de aplicaciones&#8203;:contentReference[oaicite:3]{index=3}.
